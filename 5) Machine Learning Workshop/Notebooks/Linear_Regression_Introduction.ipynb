# Introduction to Linear Regression
Linear Regression is one of the simplest and most widely used algorithms in machine learning. In this notebook, we'll explore what linear regression is, how it works conceptually, and implement it from scratch and using `scikit-learn`. No heavy math here – just the basics to help you understand its core idea!

## What is Linear Regression?
Linear regression aims to find a **line of best fit** that represents the relationship between input variables (also called **features**) and an output variable (or **target**).

The goal is to use this line to make predictions – for example, given data about house sizes, we can predict prices by fitting a line through the data.

## How Linear Regression Works (without Math)
Linear regression tries to find the best slope and intercept for a line that predicts the output values as closely as possible to the actual values. The process involves finding values that minimize the **error** – the distance between the line's predictions and the actual data points.

## Example Dataset
To understand how linear regression works, let's start by loading a simple dataset and visualizing it.


```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Generating a synthetic dataset
np.random.seed(0)
X, y = make_regression(n_samples=100, n_features=1, noise=10)

# Visualizing the dataset
plt.scatter(X, y, color='blue')
plt.title('Scatter plot of the dataset')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.show()
```

## Implementing Linear Regression from Scratch
Now, let's implement a simple linear regression model from scratch by calculating the slope and intercept for the line.

```python
# Function to calculate the slope and intercept
def linear_regression_scratch(X, y):
    n = len(X)
    X_mean = np.mean(X)
    y_mean = np.mean(y)
    
    num = sum((X - X_mean) * (y - y_mean))
    den = sum((X - X_mean) ** 2)
    
    slope = num / den
    intercept = y_mean - slope * X_mean
    return slope, intercept

# Calculating slope and intercept
slope, intercept = linear_regression_scratch(X.flatten(), y)
print(f'Slope: {slope}, Intercept: {intercept}')
```

```python
# Plotting the line of best fit from scratch
y_pred_scratch = slope * X.flatten() + intercept
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred_scratch, color='red', label='Line of Best Fit (Scratch)')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.title('Linear Regression Line (From Scratch)')
plt.show()
```

## Using Scikit-Learn for Linear Regression
Now, let's see how easy it is to implement linear regression using `scikit-learn`. We'll fit a model and compare the predictions with our from-scratch line.

```python
# Using scikit-learn's LinearRegression
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred_sklearn = model.predict(X)

# Plotting the results
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred_sklearn, color='green', label='Line of Best Fit (scikit-learn)')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.title('Linear Regression Line (scikit-learn)')
plt.show()
```

## Evaluating the Model
One way to check how well our model fits the data is to calculate the **Mean Squared Error** (MSE), which tells us the average squared difference between predictions and actual values.

```python
# Calculating Mean Squared Error for both models
mse_scratch = mean_squared_error(y, y_pred_scratch)
mse_sklearn = mean_squared_error(y, y_pred_sklearn)

print(f'Mean Squared Error (Scratch): {mse_scratch}')
print(f'Mean Squared Error (scikit-learn): {mse_sklearn}')
```
