```python
import pandas as pd
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
import numpy as np
df.drop(columns=["New_Price"], inplace=True)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
def extract_numerical(value, unit_keywords):
    if isinstance(value, str):
        for unit in unit_keywords:
            value = value.replace(unit, "")
        return pd.to_numeric(value.strip(), errors="coerce")
    return np.nan

df["Mileage"] = df["Mileage"].apply(lambda x: extract_numerical(x, ["kmpl", "km/kg"]))
df["Engine"] = df["Engine"].apply(lambda x: extract_numerical(x, ["CC"]))
df["Power"] = df["Power"].apply(lambda x: extract_numerical(x, ["bhp"]))
```

```python
df["Brand"] = df["Name"].apply(lambda x: x.split()[0] if isinstance(x, str) else np.nan)
df.drop(columns=["Name"], inplace=True)
```

```python
df.head()
```

```python
df["Mileage"].fillna(df["Mileage"].median(), inplace=True)
df["Engine"].fillna(df["Engine"].median(), inplace=True)
df["Power"].fillna(df["Power"].median(), inplace=True)
df["Seats"].fillna(df["Seats"].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

nominal_cols = ['Location', 'Fuel_Type', 'Transmission', 'Brand']
ordinal_col = ['Owner_Type']

le = LabelEncoder()
for col in nominal_cols + ordinal_col:
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.preprocessing import StandardScaler
num_cols = ['Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats']
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X = df.drop(columns=["Price"])
y = df["Price"]
```

```python
# remove outlier
X = X[(y < 1.5 * y.quantile(0.75)) & (y > 0.5 * y.quantile(0.25))]
y = y[(y < 1.5 * y.quantile(0.75)) & (y > 0.5 * y.quantile(0.25))]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rse = 1 - r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Relative Squared Error (RSE):", rse)
```
