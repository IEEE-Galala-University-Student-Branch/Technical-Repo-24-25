```python
def extract_numerical(value, unit_keywords):
    if isinstance(value, str):
        for unit in unit_keywords:
            value = value.replace(unit, "")
        return pd.to_numeric(value.strip(), errors="coerce")
    return np.nan

df["Mileage"] = df["Mileage"].apply(lambda x: extract_numerical(x, ["kmpl", "km/kg"]))
df["Engine"] = df["Engine"].apply(lambda x: extract_numerical(x, ["CC"]))
df["Power"] = df["Power"].apply(lambda x: extract_numerical(x, ["bhp"]))
```

```python
df["Brand"] = df["Name"].apply(lambda x: x.split()[0] if isinstance(x, str) else np.nan)
df.drop(columns=["Name"], inplace=True)
```

```python
df.head()
```

```python
df["Mileage"].fillna(df["Mileage"].median(), inplace=True)
df["Engine"].fillna(df["Engine"].median(), inplace=True)
df["Power"].fillna(df["Power"].median(), inplace=True)
df["Seats"].fillna(df["Seats"].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

nominal_cols = ['Location', 'Fuel_Type', 'Transmission', 'Brand']
ordinal_col = ['Owner_Type']

le = LabelEncoder()
for col in nominal_cols + ordinal_col:
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.preprocessing import StandardScaler
num_cols = ['Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats']
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X = df.drop(columns=["Price"])
y = df["Price"]
```

```python
# remove outlier
X = X[(y < 1.5 * y.quantile(0.75)) & (y > 0.5 * y.quantile(0.25))]
y = y[(y < 1.5 * y.quantile(0.75)) & (y > 0.5 * y.quantile(0.25))]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(y, bins=30, alpha=0.7, label='All Data')
plt.hist(y_train, bins=30, alpha=0.7, label='Train')
plt.hist(y_test, bins=30, alpha=0.7, label='Test')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Car Prices')
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Visualize scatter plots for all pairs of numerical features
sns.pairplot(df[num_cols])
plt.show()
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge

lambda_value = 1.0
ridge_model = Ridge(alpha=lambda_value)
ridge_model.fit(X_train, y_train)
```

```python
# Predict ridge model
y_pred = ridge_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rse = 1 - r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Relative Squared Error (RSE):", rse)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rse = 1 - r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Relative Squared Error (RSE):", rse)
```

```python
from sklearn.linear_model import LogisticRegression

# Compute bin edges for the target variable
bin_edges = np.histogram_bin_edges(y_train, bins=3)

# Bin the target variable for classification
y_train_binned = np.digitize(y_train, bin_edges, right=False)
y_test_binned = np.digitize(y_test, bin_edges, right=False)

# Train logistic regression
logistic_model = LogisticRegression(max_iter=10000)
logistic_model.fit(X_train, y_train_binned)
```

```python
from sklearn.metrics import accuracy_score

# Predict using the logistic regression model
y_pred_logistic = logistic_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test_binned, y_pred_logistic)
print("Logistic Regression Accuracy:", accuracy)
```
