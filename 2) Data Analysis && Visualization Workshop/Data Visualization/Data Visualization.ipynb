# Data Visualization with Seaborn and Matplotlib

## Section 1: Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load example dataset
df = sns.load_dataset("tips")
```

### Relational Plots

Scatter Plot:

    Description: Shows the relationship between two continuous variables.
    Usage: To identify correlations or patterns between variables.
    Example: Plotting total_bill vs. tip to see if there's a relationship between the bill amount and the tip given.

```python
# Scatter Plot
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
```

Line Plot:

    Description: Displays data points connected by straight lines, typically used for time series data.
    Usage: To visualize trends over time or continuous variables.
    Example: Plotting size vs. tip to see how tip amounts change with different party sizes.

```python
# Line Plot
sns.lineplot(data=df, x="size", y="tip")
plt.title('Line Plot: Size vs Tip')
plt.show()
```

### Categorical Plots

Bar Plot:

    Description: Shows the average value of a variable for different categories.
    Usage: To compare the average values across different groups.
    Example: Plotting day vs. total_bill to compare the average total bill for different days of the week.

```python
# Bar Plot
sns.barplot(data=df, x="day", y="total_bill")
plt.title('Bar Plot: Day vs Total Bill')
plt.show()
```

Count Plot:

    Description: Displays the count of observations in each categorical bin.
    Usage: To see the distribution of categorical data.
    Example: Counting the number of occurrences of each day.

```python
# Count Plot
sns.countplot(data=df, x="day")
plt.title('Count Plot: Day')
plt.show()
```

Box Plot:

    Description: Shows the distribution of a variable by displaying the median, quartiles, and outliers.
    Usage: To understand the spread and skewness of data.
    Example: Plotting day vs. total_bill to see the distribution of total bills across different days.

```python
# Box Plot
sns.boxplot(data=df, x="day", y="total_bill")
plt.title('Box Plot: Day vs Total Bill')
plt.show()
```

Violin Plot:

    Description: Combines aspects of the box plot and density plot to show the distribution of the data.
    Usage: To visualize the distribution and probability density of the data.
    Example: Plotting day vs. total_bill to see the distribution of total bills with density estimation.

```python
# Violin Plot
sns.violinplot(data=df, x="day", y="total_bill")
plt.title('Violin Plot: Day vs Total Bill')
plt.show()
```

Strip Plot:

    Description: Displays individual data points for a variable along a categorical axis.
    Usage: To see all observations and their distribution.
    Example: Plotting day vs. total_bill to visualize individual total bill amounts on different days.

```python
# Strip Plot
sns.stripplot(data=df, x="day", y="total_bill")
plt.title('Strip Plot: Day vs Total Bill')
plt.show()
```

### Distribution Plots

Histogram:

    Description: Shows the distribution of a single continuous variable.
    Usage: To understand the frequency distribution of data.
    Example: Plotting the distribution of total_bill amounts.

```python
# Histogram
sns.histplot(df["total_bill"], kde=True)
plt.title('Histogram: Total Bill')
plt.show()
```

Joint Plot:

    Description: Combines scatter plot and histograms to show the relationship between two variables and their distributions.
    Usage: To visualize both the correlation and individual distributions.
    Example: Plotting total_bill vs. tip to see their joint distribution.

```python
# Joint Plot
sns.jointplot(data=df, x="total_bill", y="tip", kind="scatter")
plt.title('Joint Plot: Total Bill vs Tip')
plt.show()
```

Pair Plot:

    Description: Displays pairwise relationships in a dataset by plotting each pair of variables against each other.
    Usage: To get an overview of relationships between multiple variables.
    Example: Pairwise plotting of all columns in the tips dataset.

```python
# Pair Plot
sns.pairplot(df)
plt.title('Pair Plot')
plt.show()
```

Matrix Plots

    Heatmap:
        Description: Shows data values as color in a matrix layout.
        Usage: To visualize the correlation matrix or frequency table.
        Example: Heatmap of the correlation matrix of the tips dataset.

### Matrix Plots

```python
# Heatmap
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title('Heatmap: Correlation Matrix')
plt.show()
```

## Section 2: Matplotlib

### Introduction to Matplotlib

```python
import matplotlib.pyplot as plt

# Simple Plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.title('Simple Plot')
plt.show()
```

### Plot Types

```python
# Scatter Plot
plt.scatter(x, y, color='b')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Plot Customization

```python
# Markers, Lines, Colors, Labels, and Titles
plt.plot(x, y, marker='o', linestyle='--', color='r', label='Line 1')
plt.title('Customized Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend()
plt.show()
```

```python
# Bar Plot
plt.bar(x, y, color='g')
plt.title('Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4, color='c')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

Pie Chart:

    Description: Circular chart divided into sectors to illustrate numerical proportions.
    Usage: To show part-to-whole relationships.

```python
# Pie Chart
sizes = [33, 33, 33, 99]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

### Grid Lines, Multiple Plots, and Subplots

Grid Lines:

    Description: Adds grid lines to the plot for better readability.
    Usage: To make it easier to read values off the plot.

```python
# Grid Lines
plt.plot(x, y, 'o-')
plt.title('Grid Lines')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

Multiple Plots:

    Description: Multiple lines in a single plot.
    Usage: To compare different datasets on the same axes.

```python
# Multiple Plots
plt.plot(x, y, 'o-', label='Line 1')
plt.plot(y, x, 's-', label='Line 2')
plt.title('Multiple Plots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

Subplots:

    Description: Multiple plots arranged in a grid layout within a single figure.
    Usage: To display different plots together for comparison.

```python
# Subplots
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y, 'r-')
axs[0, 0].set_title('Subplot 1')

axs[0, 1].plot(y, x, 'g-')
axs[0, 1].set_title('Subplot 2')

axs[1, 0].scatter(x, y, color='b')
axs[1, 0].set_title('Subplot 3')

axs[1, 1].bar(x, y, color='c')
axs[1, 1].set_title('Subplot 4')

fig.tight_layout()
plt.show()
```
